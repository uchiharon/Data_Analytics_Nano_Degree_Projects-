```python
# Download the prediction data frame using the request libary 
start = time.time()
url = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
respond  = requests.get(url, stream = True)
# Open the image_predictions.tsv file and save the respond from the requests.get to it
with open('image_predictions.tsv', mode = 'wb') as file:
    file.write(respond.content)
print('Request is completed. \nIt took {} seconds'.format(time.time() - start))
```

```python
df_2 = pd.read_csv('image_predictions.tsv', sep ='\t')
```

```python
df_2.head()
```

```python
# Obtain the tweet data for each tweet_id using tweeter API
# Setting up twitter api
consumer_key = 'CONSUMER_KEY'
consumer_secret = 'CONSUMER_SECRET'
access_token = 'ACCESS_TOKEN'
access_token_secret = 'ACCESS_TOKEN_SECRET'

tweet_ids = df_2.tweet_id

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)


count = 0 
fail_dict = {}
start  = time.time()

# open text file to write the requested json file to   
with open('tweet_json.txt', mode = 'w') as file:
    for tweet_id in tweet_ids:
        count += 1 
        print(str(count) + ':' + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode = 'extended')
            print('Success')
# write json file to text flie
            json.dump(tweet._json, file)
# write each json file to a new line
            file.write('\n')
        except tweepy.TweepError as e:
            print('Fail')
            fail_dict['tweet_id'] = e
            pass
end = time.time()
print(end - start)
print(fail_dict)
            
    
```

```python
# Open tweet_json.txt file and used the some of the data in the json file to from a DataFrame
tweet_dict = []
with open('tweet_json.txt') as file:
# Loop throught the each line of the file
    for files in file:
        json_file = files[:-1]
        data = json.loads(json_file) # Read the file as a json file
# Extract the desired data and save to a dictionary
        id_str = data['id_str']
        favorite_count = data['favorite_count']
        retweet_count = data['retweet_count']
        full_text = data['full_text']
        user = data['user']
        statuses_count = user['statuses_count']
        user_favourites_count = user['favourites_count']
        list_count = user['listed_count']
        friends_count = user['friends_count']
        followers_count = user['followers_count']
        tweet_dict.append({'id_str':id_str, 'favorite_count':favorite_count, 'retweet_count':retweet_count, 'statuses_count':statuses_count, 'user_favourites_count':user_favourites_count, 'list_count':list_count, 'friends_count':friends_count, 'followers_count':followers_count, 'full_text':full_text})    
file.close()     
# Make a pandas DataFrame with the created dictionary
df_3 = pd.DataFrame(tweet_dict) 
df_3.to_csv('Tweet_data.csv', index = False)
```

```python
df_3.head()
```

id_str:  The string representation of the unique identifier for this user\
favorite_count:  Tweet number of likes\
retweet:  Number of retweet\
statuses_count:  The number of Tweets (including retweets) issued by the user\
favorites_count:  The number of tweets this user has liked in the account's lifetime\
listed_count:  The number of public lists that this user is a member of\
friends_count:  The number of users this account is currently following\
followers_count:  The number of followers this account currently has\
full_text: Tweet discription

### Tideness

#### Define
Restructure DataFrames to three table named; achieve_source, twitter_a and prediction table

```python
# Drop all sources for the different table
twitter_a_clean.drop([
     'source','in_reply_to_status_id', 'in_reply_to_user_id', 'expanded_urls', 'retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp', 'text'], axis = 1, inplace = True)

predictions_clean.drop(['jpg_url'], axis = 1, inplace = True)

# Combine tables to form a single data frame
clean_data = pd.merge( twitter_a_clean, tweet_a_clean, on='tweet_id', how='inner')
combined_clean_data = pd.merge( clean_data, predictions_clean, on='tweet_id', how='inner')
```
